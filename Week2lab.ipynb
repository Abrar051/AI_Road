```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
print("Axes3D imported successfully!")

df = pd.read_csv('insurance.csv')
df.head()
```

```python
X = df.age
y = df.charges
X = X.values.reshape(-1, 1) # make it a 2-d array
X = np.hstack((np.ones_like(X), X)) # add a column of 1s to X for the intercept term
```

```python
# use gradient descent to find thetas to fit the model to the training dataset
import numpy as np
def calc_gradient(thetas, x, y):
    # y_hat = x.dot(thetas).flatten() #this version works well too. flatten a 2-d x.dot(thetas) array into a 1-d array
    y_hat = np.dot(x, thetas).flatten()
    # error = y_hat - y.flatten() # flatten y similarly. You might encounter error:
    #AttributeError: 'Series' object has no attribute 'flatten', try this:
    error = y_hat - y.values.ravel()
    gradient = (1.0/len(x)) * error.dot(x) # this calculates the gradient
    mse = np.mean(error**2) # this calculates the mean squared error
    return gradient, mse # a function can return two values

# initialisation
thetas = [0, 0] # start at this set of thetas
iter_count = 0
lr = 0.0001 # learning rate - it's a tiny value/step
tolerance = 0.0001 # accepted for convergence
mse_history = []
while True:
    iter_count = iter_count + 1
    gradient, mse = calc_gradient(thetas, X, y)
    mse_history.append(mse)
    thetas_hat = thetas - lr * gradient # this calculates the new thetas

    if np.sum(abs(thetas_hat - thetas)) < tolerance:
        print("Reach the tolerance !!!!!!")
        break

    if iter_count > 50:
        print("Hit the Max iteration !!!!!!")
        break

    thetas = thetas_hat


    if iter_count % 1 == 0:
        print("Iteration: %i - mse: %.3f" % (iter_count, mse)) # note that the cost function is half of mse
        # You can uncomment either of thetas and gradient to see what it looks like in each iteration
        print("Theta:")
        print(thetas)
        print("gradient:")
        print(gradient)


```
